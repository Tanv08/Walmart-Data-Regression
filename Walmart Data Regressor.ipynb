```python
def detect_outlier(field):
    q1 = data[field].quantile(0.25)
    q3 = data[field].quantile(0.75)
    iqr = q3-q1

    threshold = iqr*1.5
    lb = q1-threshold
    ub = q3+threshold

    outliers = [feature for feature in data[field] if feature<lb or feature>ub]
    return outliers
```

```python
outlier_features = {}
for i in column_list:
    outlier_features[i] = len(detect_outlier(i))*100/len(data)
```

Problem : We have Holiday_Flag with outlier values but its categorical feature that is why

```python
# Function to return the outlier bounds

def return_bounds(field):
    q1 = data[field].quantile(0.25)
    q3 = data[field].quantile(0.75)
    iqr = q3-q1

    threshold = iqr*1.5
    lb = q1-threshold
    ub = q3+threshold

    return [ub,lb]
```

```python
for i in outliers:
    lst = return_bounds(i)
    ub = lst[0]
    median = np.median(data[i])
    data[i] = np.where(data[i]>ub,median,data[i])
```

```python
for i in outliers:
    lst = return_bounds(i)
    lb = lst[1]
    median = np.median(data[i])
    data[i] = np.where(data[i]<lb,median,data[i])
```

```python
outlier_features = {}
for i in column_list:
    outlier_features[i] = len(detect_outlier(i))*100/len(data)
```

```python
lst = return_bounds('Unemployment')
lb = lst[1]
ub = lst[0]
```

```python
data = data[(data['Unemployment'] < ub) & (data['Unemployment'] > lb)]
```

# Feature scaling and data modelling 

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('Processed.csv')
```

```python
# Creating the target and the data separation 

x = data.drop(['Weekly_Sales'],axis=1)
y = data['Weekly_Sales']
```

```python
scaler = StandardScaler()
```

```python
scaled_data = scaler.fit_transform(x)
```

```python
# Creating the test-train split 

x_train, x_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=50)
```

# Model Training and Evaluation

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_predict = lr.predict(x_test)
```

```python
# Calculating the mean squared error

mse = mean_squared_error(y_test,y_predict)
```

```python
# Calculating root mean squared error
rmse = np.sqrt(mse)
```

```python
print('MSE',mse)
print('RMSE',rmse)
```

```python
r2_score(y_test,y_predict)
```

Inference : Linear regression is not able to explain the variance of the data because the r2_score achieved is very less

```python
# importing the model

from sklearn.tree import DecisionTreeRegressor
```

```python
tree_reg = DecisionTreeRegressor(max_depth=3)
```

```python
tree_reg.fit(x_train, y_train)

# Predict the output for the test data
y_predict = tree_reg.predict(x_test)

# Evaluate the performance of the model using mean squared error
mse = mean_squared_error(y_test, y_predict)
print("Mean squared error: ", mse)
```

```python
print(r2_score(y_test,y_predict))
```

Inference 2 : Decision Tree regressor is better than the Linear Regression in terms of the fit as the r2_score for the DTR is better than LR

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_reg = RandomForestRegressor(n_estimators=100, random_state=42)
rf_reg.fit(x_train, y_train)

# Predict the output for the test data
y_pred = rf_reg.predict(x_test)

# Evaluate the performance of the model using mean squared error and R2 score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean squared error: ", mse)
print("R2 score: ", r2)
```

Inference : Random Forest Regressor is the best fitting model for the given dataset and it was able to nearly accurate about fitting the data.
